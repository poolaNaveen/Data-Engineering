```python
import pandas as pd

# 4. List of Tuples
data4 = [
    (1, "Meera", 91),
    (2, "Sanjay", 84),
    (3, "Divya", 87)
]
df4 = pd.DataFrame(data4, columns=["ID", "Name", "Score"])
df4
```

```python
# 1. Dictionary of Lists
data1 = {
    "ID": [1, 2, 3],
    "Name": ["Aarav", "Priya", "Vikram"],
    "Score": [85, 90, 78]
}
df1 = pd.DataFrame(data1)
df1
```

```python
data7 = {
    "ID": pd.Series([1, 2, 3]),
    "Name": pd.Series(["Suresh", "Neha", "Ajay"]),
    "Score": pd.Series([88, 79, 91])
}
df7 = pd.DataFrame(data7)
df7
```
